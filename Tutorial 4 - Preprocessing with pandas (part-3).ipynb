## Extracting data from rows using loc and iloc function
#### Difference between loc and iloc is that loc finds out data by a particular value after assigning a column as index column but iloc finds out data by the default index value.

```python
import pandas as pd
```

```python
df=pd.read_csv('Advertising.csv',index_col='sales')     #Assigning 'sales' column as the index.
df
```

```python
df.loc[9.7]            #Extracting a values using it's name.
```

```python
df.loc[13.4]
```

```python
df.iloc[2]             #Using iloc and getting values at index 2.
```

```python
df.iloc[[2,5]]        #Getting values at multiple indexes using iloc function.
```

```python
df2=pd.read_csv('fatal-police-shootings-data.csv',index_col="name")      #Assigning 'name' as index column.
df2
```

```python
df2.loc["Matthew Hoffman"]         #Extracting all info about this row using loc function
```

```python
df2.loc['Nicholas Hirsh':]         #Slicing , starting from a particular value to end of dataset.
```

```python
df2.loc[:'Matthew Hoffman']           #Slicing , starting from 0th index value to a particular value in dataset.
```

```python
df2.loc[["Lewis Lee Lembke","Tim Elliot"]]          #Extracting values of two rows together.
```

```python
df2.loc["Tim Elliot":"Matthew Hoffman"]          #Extracting all values within a range.
```

```python
"Tim Cook" in df2.index             #To know if a value exists in the index or not.
```

```python
df2.iloc[1:4]            #Getting values by index in a range.
```

```python
df2.iloc[:3]          #Starting from 0th index to 3rd index.
```

```python
df2.iloc[[2,4]]        #Getting multiple values at index 2 and 4.
```

```python
df2.iloc[2:10:2]         #Starting from 0th index to 10th index with a gap of 2.
```

```python
df2.loc["John Paul Quintero","age"]             #If we want to extract value only for a particular column.
```

```python
df2.loc["Autumn Steele","date"]             #Extracting date for index "Autumn Steele".
```

```python
df2.loc["Autumn Steele",["date","age","gender","race"]]   #Extracting a row value for multiple columns.
```

```python
df2.iloc[2,4]    #Extracting age(index=4) for a row value of index 2.
```

```python
df2.iloc[2,1:5]            #Extracting value of 2nd index of row for a range(1-4) of column indexes.
```

```python
df2.iloc[7,[2,5,8]]      #Extracting values at 7th row index for multiple insexes of columns(2,5,8).
```

### Changing or replacing values. 

```python
df2.loc["Michael Rodriguez","age"]="16"    #Changing the value of age column for the row having "Michael Rodriguez" index.
df2.head()
```

```python
df2.loc["Tim Elliot",["age","city","state"]]=[32,"Paris","US"]   #Changing multiple column values for a row.
df2.head()
```

```python
copy = df2["state"]=="OR"         #Extracting data where 'state' is 'OR'.
df2[copy]
```

```python
df3=df2[copy]        #Putting this new dataframe of 88 rows into a new varibale(df3)
df3["state"]="Delhi"      #Now replacing the values of state from 'OR' to 'Delhi'
df3                #These changes are not saved in original dataset.
```

```python
df2.loc[copy]             #By using this, changes are saved in the original dataset.
```

```python
df2.loc[copy,"state"]="Delhi"          #Changing the value of state from 'OR' to 'Delhi'. 
df2                    #It has been changes permanantly in dataset.
```

## Renaming the names of columns or features 

```python
df2.rename(columns={"state":"country","city":"Indian city"})      #We are renaming two columns together using dictonaries.
```

```python
df2.rename(index={"Tim Elliot":"Tim Cook","John Paul Quintero":"John Cena"},inplace=True)
df2       #Renaming two index values by giving the index parameter and saving the changes permanantly.
```

```python
df2.loc["John Cena"]         #Extracting info of changed index to confirm the changes.
```

## Drop a column or a row 

```python
df2.drop(labels="Michael Rodriguez")     #Deleted the row with index "Michael Rodriguez".axis=0 is by default value which is for row.
```

```python
df2.drop(["Lewis Lee Lembke","Tim Cook","William Slyter"])    #Deleting multiple rows together.
```

```python
df2.drop(columns="flee",axis=1)        #Dropped the column 'flee'.axis=1 is must to apply it to column otherwise it will apply it for row.
```

```python
df2.drop(["armed","race","id"],axis=1)      #Deleted three columns together.If we don't pass axis=1 then it will return an error that falues does not exist in axis.
```

```python
city=df2.pop("city")            #Pop function is also used to delete a particular column permanantly.Here you don't need to use inplace=True.
df2         #If you apply pop function on same column twice then it will return an error as it has deleted that column permanantly in the first attempt.
```

```python
del df2["flee"]        #Also used to delete a particular column permanantly.
```
