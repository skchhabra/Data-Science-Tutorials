# How to remove a column from a dataset
### drop function is used to remove a particular column from dataset. 3 paraemeters need to be passed here :-
#### 1.Name of the column which we need to drop.
#### 2.axis=1 , it takes axis=0 by default which is used for row as we disscussed earlier.So axis=1 is very important otherwise it will assume axis to be 0 and will show an error as the column name will not be found in rows.
#### 3.inplace=True , it is very important as it makes permanent changes in dataset.If we don't pass this parameter then column will be dropped temporarily and if we will do further operations then this operation will be erased and column will be showed in dataset.

```python
df.drop("Crash",axis=1,inplace=True)      #Dropped Crash column which we added recently.
```

```python
df.head()
```

```python
df.drop(["New","Sea"],axis=1,inplace=True)          #To drop multiple columns at once.
#Remember not to run drop  function twice as it will give an error that column does not exist as it has been dropped already.
```

```python
df.head()
```

# Null values
### In our dataset , we might have null values and they can be shown as NaN or blank spaces.We need to fill them or drop them if a row is completely null before starting working on dataset.
### How to drop null values.
#### 'dropna' method is used to drop null values.

```python
df1 = pd.read_csv('fatal-police-shootings-data.csv')    #Importing new dataset.
df1
```

```python
df1.dropna(inplace=True)                  #It will drop all the rows which have null values.You can check that rows have been reduced from 5416 to 4399.
     #Inplace is required to save these changes permanantlyWe can also check other parameters of dropna by shift + TAB button.
```

```python
df1.dropna(how="any")     #how is an arguement that is passed in dropna method.Any will do the same work that is deleting all rows which have any null values.
```

```python
df1.dropna(how="all",inplace=True)     #how="all" will drop only those rows whose all values are null.This dataset doesn't have any row which has all null values.So still it has 5416 rows.
df1
```

```python
df1.dropna(axis=1)     #To drop the columns having null null values.axis = 0 is the default value which is for row.
```

### How to fill null values
#### fillna method is used to fill null values.

```python
df1          #As you can see we have dropped all rows with null values permanantly by using inplace=True so we need to import data again.
```

```python
df2 = pd.read_csv('fatal-police-shootings-data.csv') 
df2          #Now again we have our otiginal dataset with null or missing values.
```

```python
df2["age"].fillna(0,inplace=True)     #To fill value of "Age" column only with 0.
df2             #You can check alos row no.5412 has NaN for 'age' and now it has 0.
```

```python
df2["race"].fillna("W",inplace=True)           #We can fill any particular column with value of our choice.
                                 #"race" column had NaN for 5412th row and now it has 'W' which we filled
df2
```

```python
df2.tail()
```

#### Now the best way to fill null values is by filling it with mean of whole column.Doing it for age column as it has integer values.
#### We need to load the dataset again as we had set null values of 'age' as 0 permanantly using inplace=True

```python
df3 = pd.read_csv('fatal-police-shootings-data.csv')
df3.tail()         #Now as you can see we again have NaN values in 'age' column.
```

```python
df3["age"].mean()         #Mean of 'age' column
```

```python
df3.fillna(df3["age"].mean())        #Filling null values with mean of 'age' column.
```
